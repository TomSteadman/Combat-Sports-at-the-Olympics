```python
from bs4 import BeautifulSoup #better understand html
import requests 

import pandas as pd
import numpy as np
import re # regular expression
```

```python
url = 'https://en.wikipedia.org/wiki/Wrestling_at_the_Summer_Olympics'
result = requests.get(url)
soup = BeautifulSoup(result.text, 'html.parser')
```

# Mens

```python
table = soup.find_all('table', class_="wikitable")
table = table[2]
```

```python
events = ['Year']
for i in range(1,22):
  rows = table.find_all('tr')[i]
  txt = rows.find_all('td')[0].text[:-1]
  events.append(txt)
```

```python
weights = pd.DataFrame(columns = events)
```

```python
for j in range(13):
  no = 4*j+1972
  year = [no]
  for i in range(1,22):
    rows = table.find_all('tr')[i]
    try:
      txt = rows.find_all('td')[17+j].a.get('title')
      txt = txt.split(' ')[-2]
    except:
      txt = np.nan
    year.append(txt)
  year_df=pd.DataFrame([year], columns=events)
  weights = weights.append(year_df)
  
```

```python
weights
```

```python
weights.set_index('Year', inplace=True)
```

```python
weights.columns
```

```python
cols = ['light-flyweight', 'flyweight',
       'bantamweight', 'featherweight',
       'lightweight', 'welterweight',
       'middleweight', 'light-heavyweight',
       'heavyweight', 'super-heavyweight']
```

```python
gr=weights[['Greco-Roman light flyweight', 'Greco-Roman flyweight',
       'Greco-Roman bantamweight', 'Greco-Roman featherweight',
       'Greco-Roman lightweight', 'Greco-Roman welterweight',
       'Greco-Roman middleweight', 'Greco-Roman light heavyweight',
       'Greco-Roman heavyweight', 'Greco-Roman super heavyweight']]
gr.columns=cols
gr
```

```python
f=weights[['Freestyle light flyweight', 'Freestyle flyweight',
       'Freestyle bantamweight', 'Freestyle featherweight',
       'Freestyle lightweight', 'Freestyle welterweight',
       'Freestyle middleweight', 'Freestyle light heavyweight',
       'Freestyle heavyweight', 'Freestyle super heavyweight']]
f.columns=cols
```

```python
f
```

```python
gr.to_csv('men_gr_wrestling.csv')
f.to_csv('men_f_wrestling.csv')
```

# Womens

```python
table = soup.find_all('table', class_="wikitable")
table = table[3]
```

```python
events = ['Year']
for i in range(1,7):
  rows = table.find_all('tr')[i]
  txt = rows.find_all('td')[0].text[:-1]
  events.append(txt)
```

```python
events
```

```python
weights = pd.DataFrame(columns = events)
```

```python
for j in range(5):
  no = 4*j+2004
  year = [no]
  for i in range(1,7):
    rows = table.find_all('tr')[i]
    try:
      txt = rows.find_all('td')[1+j].a.get('title')
      txt = txt.split(' ')[-2]
    except:
      txt = np.nan
    year.append(txt)
  year_df=pd.DataFrame([year], columns=events)
  weights = weights.append(year_df)
```

```python
weights.set_index('Year', inplace=True)
```

```python
weights.columns = ['Freestyle flyweight', 'Freestyle bantamweight',
       'Freestyle welterweight', 'Freestyle middleweight',
       'Freestyle light-heavyweight', 'Freestyle heavyweight']
```

```python
weights.columns = weights.columns.map(lambda x: x.split(' ')[1])
```

```python
weights
```

```python
weights.to_csv('womens_wrestling_weights.csv')
```
